## The Ground Rules
* *Ignoring* possible potential energy terms, the total four-momentum of a collection of particles is the sum of the four-momenta of the particles
$$ p_\mathrm{tot} = \sum_i p_i $$
* The total four-momentum is conserved in all processes
Below we will use units where energy, momentum, and mass are all measured in energy units (eV); remember that if you want to translate to SI you will need to divide by appropriate factors of $c$ (momentum is eV/c, mass is ev/c<sup>2</sup>).

## Particle decays
* Consider the decay of a certain particle (made of a quark and an antimatter quark) called the $\pi^0$ (the *neutral pion*, generally pronounced *pi-zero*). 
    * it has a mass $m=$ 135 MeV and it is unstable with a mean (proper) lifetime of $8.5 \times 10^{-17}$ s
    * it almost always decays to two photons. 
* Let's look at the kinematics of the decay.

* The total four-momentum is conserved, and the photons have no potential energy between them
* $p_{\pi^0} = p_{\gamma_1} + p_{\gamma_2}$ as a four-vector equation
    * because this is a four-vector equation, it holds in any reference frame
* To determine details like energy or momentum individually, we need to choose a reference frame to do computations in
    * a natural choice is the one in which the $\pi^0$ is at rest. 
    * we can't choose a frame in which the photons are at rest - they are always moving at $c$!
    * in this frame,

$$
\begin{align*}
p_{\pi^0} &= [m_{\pi^0}, \mathbf{0}]\\
p_{\gamma_1} &= [E_1, \mathbf{p}_1]\\
p_{\gamma_2} &= [E_2, \mathbf{p}_2]
\end{align*}
$$

The equation $p_{\pi^0} = p_{\gamma_1} + p_{\gamma_2}$ has to hold component-by-component. This requires
$$ \mathbf{p}_1 + \mathbf{p}_2 = \mathbf{0}\\
\mathbf{p}_1 = -\mathbf{p}_2 = \mathbf{p}
$$
We then use the fact that the photon is massless to determine that $E_1 = |\mathbf{p}_1| = |\mathbf{p}|$ and $E_2 = |\mathbf{p}_2| = |-\mathbf{p}| = |\mathbf{p}|$, hence $E_1 = E_2 = E_\gamma$. Finally, we conclude $m_{\pi^0} = E_1 + E_2 = 2E_\gamma$, and therefore $E_\gamma = m_{\pi^0}/2$.

In summary: the two photons have equal energies and travel in opposite directions. Both photons have energy 67.5 MeV.

If we wanted to use some other reference frame to measure the photon energies and momenta, we could use an Lorentz transformation from the $\pi^0$ reference frame. 
* let's suppose the $\pi^0$ has energy 10 GeV in the rest frame of the experimenter ("lab frame").
* we define the direction that the $\pi^0$ is traveling to be the $x$ axis
* the angle that the photon 1 momentum makes with the $x$ axis in the $\pi^0$ rest frame is $\theta^*$ 
    * (we use the $*$ to denote the rest frame quantity).
* we want to know the $\gamma$ and the $v$ for the Lorentz transformation. The $\gamma$ factor for the $\pi^0$ can be obtained from the equation $E = \gamma m$:

$$ \gamma = 10\textrm{ GeV}/135\textrm{ MeV} = 74$$

* and then $v$ can be obtained from $\gamma$:

$$ v = \sqrt{1-1/\gamma^2} = 0.99909 $$

* ($\gamma$ factors this big mean the particles are ultrarelativistic.) The four vectors of the photons in the $\pi^0$ rest frame are (choosing coordinates so the photons are traveling in the $xy$ plane):

$$
\begin{align*}
p_{\gamma_1} &= [E_\gamma, E_\gamma \cos\theta^*, E_\gamma\sin\theta^*, 0]\\
p_{\gamma_2} &= [E_\gamma, -E_\gamma \cos\theta^*, -E_\gamma\sin\theta^*, 0]
\end{align*}
$$

* We do a Lorentz boost from the rest frame to the lab frame (velocity $-v$ along the $x$ axis), to get the following energies:

$$
\begin{align*}
E_{1,\mathrm{lab}} &= \gamma(E_\gamma + vE_\gamma\cos\theta^*)\\
&= \gamma E_\gamma(1+v\cos\theta^*)\\
&= \gamma \frac{m_{\pi^0}}{2}(1+v\cos\theta^*)\\
&= \frac{E_{\pi^0}}{2}(1+v\cos\theta^*)\\
E_{2,\mathrm{lab}} &= \frac{E_{\pi^0}}{2}(1-v\cos\theta^*)\\
\end{align*}
$$

So in general the photons will have different energies which sum to the total $\pi^0$ energy. 
* The photon traveling in the same direction as the $\pi^0$ will get an energy boost, a "blue shift", and the other will get an energy reduction, a "red shift"
* They will also generally no longer be traveling in opposite directions. In our specific example, if $\cos\theta^*=1$, then $E_{1,\mathrm{lab}} = 9.9995$ GeV, and $E_{2,\mathrm{lab}} = 0.0005$ GeV.

## Fission and Fusion
### Mass Defect
In the formula above to compute the total four-momentum of a system of particles, we ignored the possibility of potential energy. 
* In fact in many cases the potential energy binding the particles is *negative*, i.e. the total energy of the system is lower than it would be just from mass and kinetic energy.  
    * A *bound state* is one where by definition the particles don't have enough kinetic energy to escape the potential well, so the total energy of the system must be lower than the energy if the particles were completely free.
* An example of this situation is atomic nuclei
    * the protons and neutrons of a nucleus are bound together by nuclear forces
    * there is a negative energy component associated with this binding

A reduction in the total energy of a system of particles translates to a reduction in the *mass* that we compute for the system of particles.  
* For gravity and electromagnetism this effect is generally small enough to be ignored, but for nuclear forces it is noticeable. 
* For example, let's consider the helium-4 nucleus, which has two protons and two neutrons:
    * the proton mass is 1.007276 amu, the neutron mass is 1.008664 amu; so the total mass of two free protons and two free neutrons is 4.03188 amu;
    * the helium-4 nucleus mass is measured to be 4.001506 amu, which is 0.03037 amu (0.76%) smaller than the sum of the masses of the constituents.

If we were able to squish together two protons and two neutrons (at rest) to make a helium nucleus, conservation of energy implies that we need to have an extra kinetic energy floating around to ensure that energy is conserved:

$$ 2m_p + 2m_n = m_\textrm{He} + \textrm{KE}$$

Therefore, by _fusing_ protons and neutrons to make a heavier nucleus, we can create kinetic energy from stationary particles. 
* The 0.03 amu is $4.5 \times 10^{-12}$ J
    * this may seem small, but is much much bigger than e.g. the ioniziation energy of a hydrogen atom (13.6 eV = $2.2\times 10^{-18}$ J). 

We tend to refer to this as turning "mass energy into kinetic energy" although in a real sense it is better understood as getting kinetic energy from lowering the total potential energy of the protons and neutrons. 
* From the *outside*, we just measure the mass (and inertia!) of the protons/neutrons/helium, and see that the effect of the potential energy is to lower the mass of the system.

![Mass defect versus number of nucleons](images/1280px-Binding_energy_curve_-_common_isotopes.svg.png)

The difference of the of the mass of a nucleus from the total masses of the protons and neutrons (*nucleons*) in the nucleus is called the *mass defect*. 
* protons can turn into neutrons and vice versa using radioactivity (so e.g. we could turn four protons into a helium nucleus, as long as we also emitted two antielectrons and two neutrinos - this is what the sun does)
* can more or less move freely along the curve without worrying about whether we are dealing with protons or neutrons.

The nucleus with the lowest mass per nucleon is iron-56. 
* if I have nuclei lighter than iron-56, I can fuse them and get a kinetic energy release (*fusion*)
* if I have nuclei heavier than iron-56, I can split them apart and get kinetic energy (*fission*). 

From the binding energy curve it is clear that we get more energy from assembling heavier nuclei from protons and neutrons than from splitting very heavy nuclei, so hydrogen fusion is a more powerful energy source (per mass of material used) than uranium fission.

Stars take light nuclei, starting from hydrogen, and generate energy by fusing them together. 
* once stars have generated iron-56, they are not able to generate any more energy from the iron: in fact the iron can absorb energy from the star by being split apart. When a star generates an iron core, this is no longer able to generate thermal energy to support itself (and starts to be dominated by electron degeneracy pressure instead, like a metal). If the core gets too heavy, it will no longer be able to keep itself stable against gravity pulling it in, and will collapse to form a neutron star or a black hole, in the process creating a Type II supernova.

## Particle Collisions
### Particle Accelerators
We can run the particle decay process in reverse: we can collide two particles together to make a single, much heavier particle. The mass of such a particle is the mass of the system of the colliding particles (i.e. the energy in the reference frame where such a particle would be at rest, as given by conservation of momentum).

We often build accelerators to collide *antimatter* with matter. Antimatter particles are like matter particles (in particular, they have the same masses and kinds of interactions) but with opposite charges. The antimatter version of an electron, called a *positron*, has the same mass as an electron, but is positively electrically charged. When a positron collides with an electron, it has some probability of annihilating with the electron and producing a new particle or particles (photons, quarks, whatever). This makes positron-electron a great way of producing heavy unstable particles that don't normally exist.

[![Electron-positron scattering rates as a function of energy](images/ee_sigma.png)](https://pdg.lbl.gov/2020/hadronic-xsections/hadron.html)

If we have a positron with four-momentum $p_+$ and an electron with four-momentum $p_-$, the total four-momentum of a particle produced by the annihilation will be $p_+ + p_-$. The mass squared of the electron-positron system is

$$
\begin{align*}
m^2 &= (p_+ + p_-)^2\\
&= p_+^2 + 2 p_+ \cdot p_- + p_-^2\\
&= m_e^2 + 2(E_+E_- - \mathbf{p}_+ \cdot \mathbf{p}_-) + m_e^2
\end{align*}
$$

Let's consider two possible scenarios for an accelerator in the "lab frame", the reference frame of an experimentalist. One possibility is to make a positron beam with energy $E_+$ for each positron, and colliding that with stationary electrons. This gives us

$$
\begin{align*}
m^2 &= m_e^2 + 2(E_+ m_e - \mathbf{p}_+ \cdot \mathbf{0}) + m_e^2\\
&= 2m_e^2 + 2E_+ m_e\\
&= 2m_e(E_+ + m_e)
\end{align*}
$$

If we assume $E_+ \gg m_e$ (which is not so difficult to achieve) then $E_+ + m_e \approx E_+$, and $m \approx \sqrt{2m_e E_+}$. So the mass of the possible particle produced in annihilation increases with the square root of the energy of the positron beam.

The other possibility is to have opposite-direction beams of electrons and positrons (three-momenta $\mathbf{p}$ and $-\mathbf{p}$). The beam of electrons is less dense than a solid target is, so the collision probability is reduced, but let's see what happens with the mass of the particles we can produce:

$$
\begin{align*}
m^2 &= m_e^2 + 2(E_+ E_- - \mathbf{p}_+ \cdot \mathbf{p}_-) + m_e^2\\
&= 2m_e^2 + 2E_+E_- + 2|\mathbf{p}_+|^2
\end{align*}
$$

If we assume $E_+ =E_- = E_\textrm{beam} \gg m_e$ then $|\mathbf{p}_+| \approx E_+$ and we can ignore the $2m_e^2$ contribution, so 

$$ m^2 \approx 4E_+E_-\\
m \approx 2E_\textrm{beam}
$$

So the mass of the particle we can produce increases *linearly* with the energy of the beam.  For concreteness, if we have a positron beam of 10 GeV energy:
* if the electron is stationary (a *fixed target* experiment) we have $m$ = 101 MeV;
* if the electron and positrons are in counter-moving beams of equal energy (a *collider* experiment) we have $m$ = 20 GeV, nearly 200 times more.

For this reason colliders are strongly preferred over fixed target experiments if you are trying to make the heaviest possible particles.

### Compton Scattering
Light of course is a particle, so we can look into collisions of light with particles. Let's consider photons of energy $E$ colliding with a stationary electron (a process called *Compton scattering*), and investigate the kinematics of the scattered photon and electron.  We will set up the photon to initially be traveling along the $x$ axis, and let's assume it scatters in the $xy$ plane.

$$\begin{align*}
p_{\gamma, i} &= [E, E, 0, 0]\\
p_{e, i} &= [m_e, 0, 0, 0]\\
p_{\gamma, f} &= [E_f, E_f\cos\theta, E_f\sin\theta, 0]\\
p_{e, f} &= [E+m_e-E_f, E-E_f\cos\theta, -E_f\sin\theta, 0]
\end{align*}
$$

where we have already applied conservation of energy and momentum to determine the final electron four-momentum $p_{e, f}$, and forced that $p_{\gamma, f}^2 = m_\gamma^2 = 0$. We have one more condition to impose, which is that $p_{e, f}^2 = m_e^2$:

$$\begin{align*}
m_e^2 &= (E+m_e-E_f)^2 - (E-E_f\cos\theta)^2 - (-E_f\sin\theta)^2\\
&= (E-E_f)^2 + m_e^2 + 2m_e(E-E_f) - E^2 - E_f^2\cos^2\theta + 2E E_f\cos\theta - E_f^2\sin^2\theta\\
0 &= E^2 -2 E E_f + E_f^2  + 2m_e(E-E_f) - E^2 - E_f^2 + 2E E_f\cos\theta\\
&= 2m_e(E-E_f) - 2E E_f(1-\cos\theta)\\
2m_e(E-E_f) &= 2 E E_f(1-\cos\theta)\\
\frac{E-E_f}{E E_f} &= \frac{1}{E_f} - \frac{1}{E} =\frac{1-\cos\theta}{m_e}
\end{align*}
$$

In terms of the photon wavelength, $\lambda = c/\nu = 1/(E/h)$ (in our units so far, $c=1$):

$$\lambda_f - \lambda = \frac{h}{m_e}(1-\cos\theta)$$

If we put factors of $c$ back in to get SI units, we have

$$ \lambda_f - \lambda = \frac{h}{m_e c}(1-\cos\theta)$$

The factor $h/m_e c$ is called the *Compton wavelength* of the electron and is $2.426 \times 10^{-12}$ m. To observe the wavelength change from Compton scattering we want $\lambda$ to be not too much longer than this, which corresponds to x-rays. (Remember visible light is $3\sim 7 \times 10^{-7}$ m.)

Historically this was of great importance in establishing the idea of light as being composed of particles of fixed energy for a given wavelength or frequency. Classically, a free electron being shaken by passing light would only be able to re-emit light of the same frequency (therefore, wavelength), regardless of direction. Compton scattering, on the other hand, requires a frequency change of the light which depends on the angle of scattering, which is inexplicable in classical electromagnetism, but is easy to understand if the scattering makes a new photon of different energy.
