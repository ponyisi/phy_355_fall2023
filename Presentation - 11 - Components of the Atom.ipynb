Statistical mechanics and chemistry, together, pretty convincingly demonstrate that the atomic theory is right - that matter is composed of little objects called atoms, which have specific characteristics for each chemical element.  There are some interesting regularities:
* there are patterns where various elements have similar properties - the periodic table is a manifestation of this, where "similar" elements are stacked on top of each other (e.g., alkali metals are soft and extremely reactive, and are the "next heaviest" elements after noble gases, which are extremely unreactive;
* the mass of atoms of various elements are often close to integer multiples of the mass of an atom of hydrogen;
* electrochemistry tells us that electric charge can be transferred between various compounds, and the amount of charge that is transferred per mole of substance is a multiple of the "Faraday constant," 96,500 C/mol;
* atoms emit and absorb light at specific (and the same) wavelengths. When the light from atoms is spread out into a spectrum, this leads to "lines".

![Periodic table](images/800px-Simple_Periodic_Table_Chart-en.svg.png)

It is tempting to try to explain these patterns as arising from some kind of internal structure of atoms.

## Cathode Rays & Electrons

In the 1870s William Crookes was able to make a very good vacuum in glass tubes ("Crookes tubes"). Positive and negative (anode and cathode) electrical terminals could be put in the tubes, with a large electric potential between them. As the vacuum was made better, it became apparent that something was being emitted from the negative terminals and traveling in more or less straight lines towards the positive terminals; when they overshot and hit the glass, they produced a glow.  Because the particles came from the negative terminal, they were called "cathode rays."  Their motion was bent by a magnetic field, which showed that they had negative charge.  People initially thought they might be charged atoms, or some new kind of electromagnetic phenomenon.

The most famous name associated with the study of cathode rays was J. J. Thomson. In one of his experiments, he was able to measure the ratio of the mass and the charge of the objects in the cathode rays.  Let's see how that worked.  First, objects with a charge $q$ moving with velocity $\vec{v}$ in a magnetic field $\vec{B}$ experience a force perpendicular to the motion and the magnetic field:

$$ \vec{F} = q \vec{v} \times \vec{B} $$

Since the force is always perpendicular to the velocity, the effect is to provide a centripetal force to the particle. If  $\vec{v}$ and $\vec{B}$ are perpendicular, then $\vec{F}$ will cause the particle to move in a circle, and standard equations from mechanics give us the radius of the circle:

$$ \begin{align*} F &= \frac{mv^2}{r} = qvB\\
r &= \frac{mv}{qB} = \frac{m}{q}\frac{v}{B} \end{align*}$$

The experimentalist gets to choose the size of the magnetic field $B$, so if we are able to measure the speed $v$ we can get $m/q$.

We can measure $v$ by carefully balancing the force from the magnetic field with another force from an _electric_ field perpendicular to the motion of the cathode rays. The magnitude of the electric force on the particle is $qE$, without any dependence on speed.  If the two forces balance, we must have

$$ \begin{align*}
qvB &= qE\\
v &= \frac{E}{B}
\end{align*}
$$

hence

$$ \frac{m}{q} = r \frac{B}{v} = \frac{r B^2}{E} $$

This yields a value &approx; 5.7 &times; 10<sup>-12</sup> kg/C.  The same value is obtained for all cathode rays, regardless of what the material of the cathode is, and the conclusion is that all cathode rays are the same kind of object.  A value of $m/q$ could also be extracted for *ions*, because one can determine how much mass of a material one needs to extract a coulomb of charge in an electrochemical reaction. For hydrogen, this number is 1.04 &times; 10<sup>-8</sup>.  For various reasons Thomson concluded that the charge of cathode rays was the same as the charge of a hydrogen ion, which leads to the conclusion that cathode rays have mass $(5.7 \times 10^{-12})/(1.04 \times 10^{-8}) = 5.4 \times 10^{-4}$ times that of a hydrogen atom. (This is often seen in the reciprocal, so a hydrogen atom is 1837 times heavier than a cathode ray particle.)  This meant that the cathode rays were some new kind of non-atomic particle, which got named "electrons."

Thomson drew the (correct) conclusion that electrons were a constituent of atoms, though he had no real picture for what the  distribution would be; he also didn't have any evidence for how the positive charge of the atoms would be distributed.

## Radioactivity
The turn of the 20th century featured the discovery of many kinds of "rays" beyond cathode rays.  The general sense of a "ray" is something that seems to travel as a "particle" - located in a small area, travels in a straight line unless forces are acting on it.
* X-rays: produced by stopping electrons in material
* &alpha;, &beta;, &gamma; rays: produced in materials containing various elements. 
   * Identified as being different from each other by how far they travel through material, and that &alpha; and &beta; rays are deflected by magnetic fields while &gamma; are not
   * Experimental results from Rutherford and Becquerel showed that &alpha; rays had the same $m/e$ as doubly-ionized helium, while &beta; rays had the same $m/q$ as electrons (although could be positively charged as well).
https://en.wikipedia.org/wiki/Radiation#/media/File:Alfa_beta_gamma_radiation_penetration.svg

Different elements have different characteristic types of ray emission (some have none at all, some are primarily &alpha; emitters, etc.), and the emission is termed "radioactivity." Elements with atomic numbers above lead are all radioactive.

https://en.wikipedia.org/wiki/Radioactive_decay#/media/File:Periodic_Table_Stability_&_Radioactivity.svg

X-rays and &gamma; rays are now understood to be very high frequency electromagnetic radiation. &alpha; particles are indeed doubly-ionized helium, while &beta; are electrons or their antimatter partners (positrons).

## Nuclei
In experiments from 1908-1913 in which &alpha; rays were sent through a very thin gold foil, Rutherford's group (in particular the students Geiger and Marsden) demonstrated that sometimes the &alpha; particles were deflected by extremely large angles, up to 180 degrees. 

* impossible to explain if one assumed the mass of an atom was uniformly smeared over the volume of an atom (as in a model of Thomson)
* implied instead that the gold atoms had structure, with a very small component (compared to the full size of the atom) much heavier than the &alpha; particles

Rutherford termed this heavy, positively charged component the "nucleus" of the atom; the electrons were then assumed to orbit around the nucleus in some fashion (presumably due to their electrostatic attraction).  

Moseley showed (1914) that an element's *atomic number*, i.e. its position in the periodic table, was also the electric charge of the nucleus, thus suggesting that different elements had different (integer) numbers of some positively-charged object in the nucleus.

https://en.wikipedia.org/wiki/Ernest_Rutherford#/media/File:Gold_foil_experiment_conclusions.svg


## Mutability of atoms
* Ancient Greek notion of "atoms": indivisible, unchanging chunks 
   * by definition alchemy couldn't be real, because an atom of one type couldn't turn into another
* Rutherford's group saw that radioactive thorium turns into radium, demonstrating that it was possible for one kind of atom to turn into another
* &alpha; and &beta; particle emission is associated with a change of one element to another (while &gamma; radiation, on its own, is not).
* The reverse process, _artificial_ nuclear transmutation, was first achieved in 1925, by bombarding nitrogen with alpha particles and turning it into oxygen.  (For curiosity: it turns out to be much easier to artificially turn gold into lead than the other way around.)

## Protons and Neutrons
* In the &alpha;-N collisions, it was observed that hydrogen was produced. A guess (which turned out to be correct) was that hydrogen was somehow present in all nuclei
   * the positively-charged nucleus of hydrogen itself was called the *proton* under the assumption that hydrogen, with atomic number 1, contained a single basic object.

*Atomic weight* ("A"), in terms of the mass of hydrogen, is larger than *atomic number* ("Z"), so it was clear that there were more particles in the nucleus than just the bare number of protons needed to get the correct electric charge. 
* First idea: there were A protons inside a nucleus, with a balancing number A-Z of electrons included to get the correct charge.  (How exactly the electrons were kept inside wasn't clear, and there wasn't an obvious sign of them.)  
* Better observation: In collisions of alpha particles with beryllium, James Chadwick (1932) observed the production of a new kind of ray (not like &alpha;, &beta;, or &gamma;) that was neutral and had mass around that of the proton.  This is called the neutron.
   * It was quickly accepted that in fact nuclei are composed of Z protons and A-Z neutrons.

There can be several different variants ("isotopes") of the same element, which differ in their atomic weights but have the same electric charge. This is interpreted as having the same number of protons but a different number of neutrons for each isotope.
* The atomic weight for each isotope is very close to an integer multiple of the weight of hydrogen
* when the _average_ atomic weight for an element isn't close to an integer, it's because we're seeing a mixture of the different isotopes in nature.  

At this point the fundamental particles making up atoms were known: protons and neutrons, together in a small nucleus containing most of the atom's mass, and electrons surrounding them.

## Atomic Spectra
In classical electromagnetism, any accelerating charged object will radiate electromagnetic waves.  
* Classically an electron in orbit around a nucleus should radiate energy via electromagnetic waves, with a frequency that depends on the frequency of the orbit (which could be anything).  

So atoms ought to collapse, and they ought to release a continuous spectrum of light while doing so.  Neither of these things is true.

In reality, atoms don't collapse, and elements and compunds emit and absorb light at very specific wavelengths. These are so regular that they can be used to determine the chemical composition of things just by looking at them; for example, this is how we study the elements present in the Sun and other stars.

https://en.wikipedia.org/wiki/Hydrogen_spectral_series#/media/File:Hydrogen_spectrum.svg

https://en.wikipedia.org/wiki/Emission_spectrum#/media/File:Emission_spectrum-Fe.svg

https://physics.nist.gov/PhysRefData/ASD/lines_form.html

Emission and absorption occur at the same wavelengths for a given chemical.  They generally have a somewhat complex form, but hydrogen has a surprisingly simple spectrum: there are several "series" of closely-spaced lines, with something of a gap between them.  The wavelengths follow a very simple form, observed by Balmer (for his namesake series) and generalized by Rydberg:

$$ \frac{1}{\lambda} = R_H \left(\frac{1}{n_1^2}-\frac{1}{n_2^2}\right) $$

where $R_H$ is the "Rydberg constant" &approx; 0.011 nm<sup>-1</sup>. So, for example, the "Lyman-&alpha;" line corresponds to $n_1 =1$ and $n_2=2$:

$$ \frac{1}{\lambda_{\mathrm{Ly-}\alpha}} = 0.011\textrm{ nm}^{-1}\left(1 - \frac{1}{4}\right) = $$
$$ \lambda_{\mathrm{Ly-}\alpha} = 121\textrm {nm}$$

If we recall that Planck and Einstein suggested that the energy of light quanta was related to their frequency $\nu$ via $E = h\nu$, and that $\nu = c/\lambda$, this suggests that the energy of the emitted light (and thus, presumably, the change in the electron energy) has a simple pattern

$$ E = h\nu = hc/\lambda = hc R_H \left(\frac{1}{n_1^2}-\frac{1}{n_2^2}\right) $$

where $hc R_H = 2.18 \times 10^{-18}$ J = 13.6 electronvolts (eV).

## Sort-of-quantum: Bohr model

Niels Bohr introduced a model for the atom that was able to reproduce the energy spectrum for hydrogen above. 
* The idea is that the energy levels were set by the electrons orbiting at certain fixed distances from the nucleus, attracted by the electric field of the nucleus
   * The closer the electrons are to the nucleus, the _lower_ (more negative) their electrical potential energy will be 
* The rule that Bohr stumbled on to generate the energy levels properly was to assert that the _orbital angular momentum_ of the electrons needed to be integer multiples of $h/2\pi$

The velocity of a classical particle in a circular orbit in a $1/r^2$ force depends on the radius of the orbit (e.g. Kepler's Third Law), so for a single classical electron orbiting a proton we would have a velocity

$$ v = \sqrt{\frac{e^2}{4\pi\epsilon_0 m_e r}} $$

where $e$ is the charge on the electron, $\epsilon_0$ is the permittivity of free space, $m_e$ is the electron mass, and $r$ is the radius of the orbit.  We also have

$$ E_\textrm{total} = E_\textrm{potential} + E_\textrm{kinetic} = -\frac{1}{2}m_e v^2 = -\frac{1}{8\pi\epsilon_0 r}$$

Imposing the condition that

$$ m_e vr = nh/2\pi$$

with $n$ an integer forces the following radii:

$$ r_n = \frac{\epsilon_0 n^2 h^2}{\pi e^2 m_e} $$

and

$$ E_n = -\frac{e^4 m_e}{8 n^2 h^2 \epsilon_0^2} = \frac{-13.6\textrm{ eV}}{n^2} $$
